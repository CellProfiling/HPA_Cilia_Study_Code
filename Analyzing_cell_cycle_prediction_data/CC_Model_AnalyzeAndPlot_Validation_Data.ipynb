```python
# %% Import packages

import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kruskal
import scikit_posthocs as sp
from statannotations.Annotator import Annotator

datapath = 'E:\\CP\Pub\\2024 Cilia Atlas\\Data\\2025_CC_predict\\'
# datapath = '/home/jan/Documents/Cilia_Atlas_Data/2025_CC_predict/'

output_folder = os.path.join(datapath,"Validation", "Plots")
```

```python
# %%

# List files to load
csv_files = glob.glob(os.path.join(datapath,"Validation","intensity_csv_files","*_intensity.csv"))
print(csv_files)
```

```python
# %%

# Load files and concatenate
for filepath in csv_files:
    # Get gene name
    gene_name = os.path.splitext(os.path.basename(filepath))[0]
    gene_name = gene_name.split("_")[0]

    print("############################\nAnalyzing prediction data for " + gene_name)

    # Read the CSV file
    df = pd.read_csv(filepath)
    df.columns = [col.strip() for col in df.columns]
    df.columns = [col.replace(" ", "_") for col in df.columns]
    df.columns = [col.replace("GMMN", "GMNN") for col in df.columns]
    
    # Extract the file name without its directory path
    filename = os.path.basename(filepath)   # e.g., "CCNA2_intensity.csv" 

    # Optionally remove the extension (e.g., ".csv")
    filename_no_ext = os.path.splitext(filename)[0] # e.g., "CCNA2_intensity"

    # Add a column indicating the source file (or a cleaned up version of it)
    df["source"] = filename_no_ext

    #####################
    ## CREATE FIGURES
    #####################

    # NORMALIZE Nuclear intensity values of the gene of interest

    quantM = 0.99
    quantMText = "q" + str(quantM).replace(".", "")
    
    lower_q = df["measured_expression"].quantile(1-quantM)
    upper_q = df["measured_expression"].quantile(quantM)

    # Create a new column "norm_predicted_GMNN" by linear normalization
    df["norm_measured_exp_"+quantMText] = (df["measured_expression"] - lower_q) / (upper_q - lower_q)

    print('max norm measured exp ' + str(max(df["norm_measured_exp_"+quantMText])))
    print('min norm measured exp ' + str(min(df["norm_measured_exp_"+quantMText])))

    # NORMALIZE GMNN Prediction values

    quant = 0.99
    quantText = "q" + str(quant).replace(".", "")

    lower_q = df["predicted_GMNN"].quantile(1-quant)
    upper_q = df["predicted_GMNN"].quantile(quant)

    # Create a new column "norm_predicted_GMNN" by linear normalization
    df["norm_predicted_GMNN_"+quantText] = (df["predicted_GMNN"] - lower_q) / (upper_q - lower_q)

    # Bin the normalized values into four categories
    bins = [-np.inf, 0.25, 0.5, 0.75, np.inf]
    labels = ["<0.25", "0.25-0.5", "0.5-0.75", ">0.75"]
    df["binned_norm_GMNN_"+quantText] = pd.cut(df["norm_predicted_GMNN_"+quantText], bins=bins, labels=labels)

    print("\nBin results for " + gene_name)
    print(df["binned_norm_GMNN_"+quantText].value_counts(sort=False))


    #####################################
    # Apply statistical testing
    # Get columns:
    group_col = f"binned_norm_GMNN_{quantText}"
    value_col = f"norm_measured_exp_{quantMText}"

    # Overall nonparametric test: Kruskal-Wallis 
    # - Checks if there's a statistically significant difference among all bins
    groups = [group_df[value_col].dropna() for _, group_df in df.groupby(group_col,observed=False)]
    if len(groups) > 1: # Ensure we actually have multiple groups 
        H, p_overall = kruskal(*groups)
        print(f"Kruskal-Wallis result for {filename_no_ext}: H={H:.3f}, p={p_overall:.3e}")

    # Post-hoc pairwise test using Dunnâ€™s test (adjust p-values with Bonferroni/Holm, etc.)
    # scikit-posthocs returns a matrix of p-values for each group comparison
    dunn_results = sp.posthoc_dunn( df, val_col=value_col, group_col=group_col, p_adjust="bonferroni" )
    print("\nPairwise Dunnâ€™s test p-values:\n", dunn_results)

    plt.figure(figsize=(3, 2))
          
    ax = sns.violinplot( data=df, x="binned_norm_GMNN_"+quantText, 
                y="norm_measured_exp_"+quantMText,
                # cut=0, # By default, extends beyond extreme values; cut=0 keeps it within observed data range
                alpha = 0.5)
        
    plt.xlabel("predicted nuclear GMNN intensity\n(percentile normalized" + quantText + ")")
    plt.ylabel("measured nuclear " + gene_name + " \n(percentile normalized " + quantMText + ")")
    plt.ylim(-0.5,1.7)

    # Annotate significance on plot with statannotations
    # For each pair of groups, fetch the p-value from the post-hoc results matrix 
    group_labels = ( df[group_col].cat.categories.tolist() if isinstance(df[group_col].dtype, pd.CategoricalDtype) else sorted(df[group_col].unique()) )

    # If group_labels are strings, ensure they match how they appear on the x-axis 
    pairs = []
    for i in range(len(group_labels)):
        for j in range(i+1, len(group_labels)):
            pairs.append((group_labels[i], group_labels[j]))

    # Convert table of p-values to a dict:
    p_values = {}
    for i, g1 in enumerate(group_labels):
        for j, g2 in enumerate(group_labels): 
            # Avoid repeating diagonal or lower half 
            if j <= i:
                continue
            p_val = dunn_results.loc[g1, g2]
            p_values[(g1, g2)] = p_val

    
    print("\nP Values Summary:")
    for (group1, group2), pval in p_values.items():
        print(f"{group1} vs {group2}: p = {pval:.2e}")

    # Add annotations
    annotator = Annotator(ax, pairs, data=df, x=group_col, y=value_col)

    # We'll supply the p-values we already have, instead of letting statannotations compute them
    annotator.configure( test=None, # we already have p-values
                        text_format="star", # use stars: "", "", "", etc. 
                        loc="outside", verbose=False )
    annotator.set_pvalues_and_annotate([p_values[pair] for pair in pairs])

    violinplot_path = os.path.join(output_folder, filename_no_ext + "_violinplot_dblnorm_stat_" + quantText + "")
    plt.savefig(violinplot_path + ".png", dpi=300, bbox_inches="tight")
    plt.savefig(violinplot_path + ".svg", dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()

    #####################################
    plt.figure(figsize=(3, 2))
    # Optionally overlay a stripplot or boxplot for more detail, e.g.:        
    ax = sns.violinplot( data=df, x="binned_norm_GMNN_"+quantText, 
                y="norm_measured_exp_"+quantMText,
                # cut=0, # By default, extends beyond extreme values; cut=0 keeps it within observed data range
                alpha = 0.5)

    plt.xlabel("predicted nuclear GMNN intensity\n(percentile normalized" + quantText + ")")
    plt.ylabel("measured nuclear " + gene_name + " \n(percentile normalized " + quantMText + ")")
    plt.ylim(-0.5,1.7)

    # Add annotations
    annotator = Annotator(ax, pairs, data=df, x=group_col, y=value_col)

    # We'll supply the p-values we already have, instead of letting statannotations compute them
    annotator.configure( test=None, # we already have p-values
                        text_format="simple", # simple, star, full
                        loc="outside", verbose=False )
    annotator.set_pvalues_and_annotate([p_values[pair] for pair in pairs])

    violinplot_path = os.path.join(output_folder, filename_no_ext + "_violinplot_dblnorm_stat2_" + quantText + "")
    plt.savefig(violinplot_path + ".png", dpi=300, bbox_inches="tight")
    plt.savefig(violinplot_path + ".svg", dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()
```
